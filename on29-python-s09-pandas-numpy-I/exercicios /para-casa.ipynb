```python
import pandas as pd
df = pd.read_csv('data_sample_DF.csv', sep=';', encoding='latin-1', on_bad_lines='skip')
df
```

Calcular a média da temperatura da amostra

```python
df = df.dropna(subset=['TEMPERATURA DO AR - BULBO SECO. HORARIA (C)'])
df_media = df['TEMPERATURA DO AR - BULBO SECO. HORARIA (C)']
df_media.mean()
```

Retirar nulos da coluna 'RADIACAO GLOBAL (Kj/m2)'

```python
df = df.dropna(subset=['RADIACAO GLOBAL (Kj/m2)'])
df.shape
```

Copiar o dataframe reduzindo para 3 colunas (a sua escolha) e 1000 linhas (aleatórias)

```python
df_reduzido = df[['Data', 'ESTACAO', 'PRECIPITAO TOTAL. HORARIO (mm)', ]]
df_reduzido.sample(1000)
```

Normalizar coluna (qualquer uma)

```python
df_normalizado = df.copy()
coluna = 'TEMPERATURA DO AR - BULBO SECO. HORARIA (C)'
df_normalizado[coluna] = (df_normalizado[coluna] - df_normalizado[coluna].min()) / (df_normalizado[coluna].max() - df_normalizado[coluna].min())
df_normalizado[coluna]
```
