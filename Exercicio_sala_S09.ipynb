```python
import pandas as pd
```

```python
df = pd.read_csv('data_sample_PB.csv', sep=';')
```

```python
df
```

1- Calcular média da temperatura da amostra:

```python
media_temperatura_PB = df['TEMPERATURA DO PONTO DE ORVALHO (C)'].mean()
media_temperatura_PB
```

2- Retirar as linhas nulas da coluna RADIACAO GLOBAL (Kj/m2)

```python
df = df.dropna(subset=['RADIACAO GLOBAL (Kj/m2)'])
df
```

3- Reduzir o df para 3 colunas e 1.000 linhas (aleatórias) da amastra da PB

```python
amostra = df[['PRECIPITAO TOTAL. HORARIO (mm)', 'TEMPERATURA DO PONTO DE ORVALHO (C)', 'ESTACAO']].sample(1000)
amostra
```

4- Normalize uma coluna

```python
df_normalizado = df.copy()

coluna = ['RADIACAO GLOBAL (Kj/m2)']
df_normalizado[coluna] = (df_normalizado[coluna] - df_normalizado[coluna].min()) / (df_normalizado[coluna].max() - df_normalizado[coluna].min())
df_normalizado[coluna].plot()
```
