**EXERCÍCIO 1:**
MÉDIA - Calcular a média da temperatura da amostra.




```python
import pandas as pd
```

```python
df = pd.read_csv('data_sample_RJ.csv', sep=';')
```

```python
df
```

```python
temperatura = ['TEMPERATURA DO AR - BULBO SECO. HORARIA (C)',
           'TEMPERATURA DO PONTO DE ORVALHO (C)',
           'TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)',
           'TEMPERATURA MINIMA NA HORA ANT. (AUT) (C)',
           'TEMPERATURA ORVALHO MAX. NA HORA ANT. (AUT) (C)',
           'TEMPERATURA ORVALHO MIN. NA HORA ANT. (AUT) (C)']

media_temperatura = df[temperatura].mean()
```

```python
media_temperatura
```

**EXERCÍCIO 4:**
NORMALIZAR - normalizar qualquer coluna


```python
import matplotlib.pyplot as plt
```

```python
coluna = ['VENTO. DIREAO HORARIA (gr) ((gr))']

df[coluna] = (df[coluna] - df[coluna].min())/(df[coluna].max() - df[coluna].min())
```

```python
coluna = ['VENTO. VELOCIDADE HORARIA (m/s)']

df[coluna] = (df[coluna] - df[coluna].min())/(df[coluna].max() - df[coluna].min())
```

```python
df[['VENTO. DIREAO HORARIA (gr) ((gr))','VENTO. VELOCIDADE HORARIA (m/s)']].plot()

plt.show()
```
