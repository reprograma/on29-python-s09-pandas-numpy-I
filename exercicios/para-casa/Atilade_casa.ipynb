```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_sample_RN.csv', sep=';' , encoding='latin-1', on_bad_lines='skip')
df
```

```python
media = df['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)'].mean()
print(f"A media da temperatura maxima {media:.2f}")
```

```python
df = df.dropna(subset = ['RADIACAO GLOBAL (Kj/m2)'])
df['RADIACAO GLOBAL (Kj/m2)']
```

```python
colunas = ['RADIACAO GLOBAL (Kj/m2)', 'TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)', 'VENTO. RAJADA MAXIMA (m/s)']
mine_df = df[colunas].sample(n=1000,random_state=1)
mine_df
```

```python
df_normalizado = df.copy()
df_normalizado = df_normalizado.dropna(subset=['VENTO. RAJADA MAXIMA (m/s)'])
ventos = "VENTO. RAJADA MAXIMA (m/s)"
df_normalizado[ventos]
df_normalizado[ventos] = (df_normalizado[ventos] - df_normalizado[ventos].min()) / (df_normalizado[ventos].max() - df_normalizado[ventos].min())
df_normalizado[ventos]
```
