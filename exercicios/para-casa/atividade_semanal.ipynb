**Importar** a biblioteca "pandas":

```python
import pandas as pd
```

**.read_csv -** Leitura do databasedataframe:

```python
df = pd.read_csv('data_sample_SP.csv', sep = ';')
```

```python
df
```

**.mean -** Média da Temperatura do Ar - Bulbo Seco:

```python
df['TEMPERATURA DO AR - BULBO SECO. HORARIA (C)'].mean()
```

**.dropna -** Retirada de NaN da Radiação Global (Kj/m²):

```python
df.dropna(subset = ['RADIACAO GLOBAL (Kj/m2)'])
```

**.sample -** Seleção de 3 colunas e 1000 colunas para um dataframe reduzido (df_reduzido):

```python
df_reduzido = df[['ESTACAO', 'DATA DE FUNDACAO', 'TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)']].sample(1000)
df_reduzido = df_reduzido.reset_index(drop=True)
```

```python
df_reduzido
```

**Normalização de colunas -** converter range para uma mesma base, entre 0 e 1, a fim de plotar posteriormente:

```python
coluna1 = 'PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO. HORARIA (mB)'
df[coluna1] = (df[coluna1] - df[coluna1].min()) / (df[coluna1].max() - df[coluna1].min())
```

```python
coluna = 'VENTO. RAJADA MAXIMA (m/s)'
df[coluna] = (df[coluna] - df[coluna].min()) / (df[coluna].max() - df[coluna].min())
```

** Importar matplotlib -** Plotar no gráfico:

```python
import matplotlib.pyplot as plt
df[['PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO. HORARIA (mB)', 'VENTO. RAJADA MAXIMA (m/s)']].plot()
```

**.sort_values -** altera a ordem da coluna de acordo com a orientação - "ascending = True (ordem crescente)", "descending = True (ordem decrescente)":

```python
df.sort_values(by = 'RADIACAO GLOBAL (Kj/m2)', ascending = True)
```
