```python
import pandas as pd # importando

df = pd.read_csv('data_sample_AM.csv', sep=';')
df
```

```python
# calcular a média da temperatura da amostra

temperatura = ['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)', 'TEMPERATURA MINIMA NA HORA ANT. (AUT) (C)'] # recebendo as colunas que serão manipuladas

media_temp_max = df['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)'].mean() # calculando a média da temperatura maxima
media_temp_min = df['TEMPERATURA MINIMA NA HORA ANT. (AUT) (C)'].mean() # calculando a média da temperatura mínima

media_geral = media_temp_max + media_temp_min # somando a média das temperaturas
print(f'A média das temperaturas é: {media_geral:.2f}°')
```

```python
# retirar nulos da coluna 'RADIACAO GLOBAL (Kj/m2)' - aproveitei o ensejo para retirar das outras colunas que vou usar posteriormente também
df[['RADIACAO GLOBAL (Kj/m2)', 'TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)', 'TEMPERATURA MINIMA NA HORA ANT. (AUT) (C)']] = df[['RADIACAO GLOBAL (Kj/m2)', 'TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)', 'TEMPERATURA MINIMA NA HORA ANT. (AUT) (C)']].fillna(0)
```

```python
# copiar o dataframe reduzindo para 3 colunas (a sua escolha) e 1000 linhas (aleatórias)
df_colunas = df[['RADIACAO GLOBAL (Kj/m2)', 'TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)', 'TEMPERATURA MINIMA NA HORA ANT. (AUT) (C)']].sample(1000)
df_colunas
```

```python
# Bônus: normalizar coluna (qualquer uma)

df_normalizado = df.copy()

coluna = "TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)"

df_normalizado['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)'] = (df_normalizado['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)'] - df_normalizado['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)'].min()) / (df_normalizado['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)'].max() - df_normalizado['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)'].min())

df_normalizado['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)']
```

```python
# Bônus II: pesquisar sobre outras formas de processamento de dados além das vistas em sala de aula

#groupby

# funções de pandas pedem que as datas de um df esteja no formato date time para serem manipuladas, como não tenho certeza, vou forçar a conversão
df['Data'] = pd.to_datetime(df['Data'])

# Agrupar por data e calcular estatísticas
df_agrup = df.groupby(df['Data'].dt.strftime('%d/%m/%Y'))['TEMPERATURA MAXIMA NA HORA ANT. (AUT) (C)'].agg(['mean', 'max', 'min']) # a função agg (agregate) aplica agragação de funções como as usadas para media, maxima e minima


# renomeação das colunas (se não renomear, output será mean, max e min)
df_agrup.columns = ['Temperatura Média', 'Temperatura Máxima', 'Temperatura Mínima']

print(f"\nEstatísticas de Temperatura do Amazonas por data:\n{df_agrup}")
```
